# Classic Time Series Decomposition -> 1920 <br>
Teknik untuk memisahkan time series menjadi trend, seasonal dan residual menggunakan moving average, ada 2 tipe: <br>
- Additive (Penjumlahan) = Trend + Seasonal + Residual <br> 
- Multiplicative (Perkalian) = Trend * Seasonal * Residual <br>

<br>
Additive dipakai pada <b>Trend dan Seasonal yang tidak terlalu bervariasi</b> <br>
Multiplicative dipakai pada <b>trend dam seasonal yang berubah seiring berjalannya waktu</b><br>
<br>
Trend = kemiringan naik atau turun suatu time series pada suatu periode tertentu<br>
Seasonal = Pola unik yang terlihat di suatu interval waktu karena faktor musiman<br>
Residual atau Error = Variasi dari data yang tidak dapat dijelaskan<br>

Contoh:<br>
- Additive = Terdapat peningkatan pelanggan 10.000 orang / tahun<br>
- Multiplicative = Terdapat peningkatan pelanggan 10% / tahun

# Seasonal vs Cyclical<br>
<b>seasonal = periodenya sudah tetap, urutannya selalu sama, polanya selalu sama</b><br>
<b>cyclical = kadang naik kadang turun, periodenya gaselalu sama, polanya selalu beda dan sulit diprediksi</b><br><br>
sebagai catatan tambahan antara seasonal dan cyclical<br>
- seasonal = berulang<br>
- cyclical = tidak berulang, bergerak naik turun secara aperiodik(tanpa periode yang tetap) contohnya harga saham<br>
